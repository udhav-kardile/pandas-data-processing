# Merging of Two DataFrames --------------> DB mgmt operations

```python
df1 = pd.read_excel('CDAC_DataBook.xlsx',sheet_name='Pune')
df1.head()
```

```python
df2 = pd.read_excel("CDAC_DataBook.xlsx",sheet_name="Mumbai")
df2.head()
```

```python
pd.merge(df1,df2)            # since name,subj, grade has no common record to merge
```

```python
pd.merge(df1,df2,on='Subject')          # returns only those sub which are common in both DF
```

```python
pd.merge(df1,df2, on=['Subject','Grade'])         # returns records where subject + grades are common
```

```python
pd.merge(df1,df2, on='Subject',how='inner')         # only common subjects from both are returned ---> INTERSECTION
```

```python
pd.merge(df1,df2, on='Subject',how='outer')          # common,uncommon subjects both are returned ---> UNION
```

```python
df1 = pd.read_excel("CDAC_DataBook.xlsx",sheet_name='Salary')
df1.head()
```

```python
pd.merge(df1,df2,how='outer')           # select all, but common are returned once ---> union
```

```python
pd.merge(df1,df2,how='left')          # returns all records from df1, and commons from df2
```

```python
pd.merge(df1,df2,how='right')
```

# Wide Structure & Broad Structure

```python
df = pd.read_excel('CDAC_DataBook.xlsx',sheet_name='Health')
df
```

```python
# 1 column for year combinely(2019,2020,2021)
df1 = df.melt(id_vars=['Country','Disease'],var_name='Year',value_name='Cases') # converting broad str to Tall str
df1
```

```python
df.shape
```

```python
df1.shape     # broad str ----> rows are more, cols are less
```

# pivot table
- Converting Tall str to Broad str

```python
df2 = df1.pivot_table(index=['Country','Year'],columns = 'Disease',values='Cases').reset_index()
df2
```

```python
df1.pivot_table(index =['Country','Year'],columns='Disease',values='Cases').reset_index().rename_axis(columns=None)
```
